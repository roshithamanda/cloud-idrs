## Objectives:
- Understand dataset structure
- Analyze attack distribution
- Visualize feature correlations
- Identify class imbalance
- Prepare for preprocessing

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for better visuals
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

print("✅ Libraries imported successfully!")
```

## 1. Load NSL-KDD Dataset

```python
# NSL-KDD column names
columns = [
    'duration', 'protocol_type', 'service', 'flag', 'src_bytes', 'dst_bytes',
    'land', 'wrong_fragment', 'urgent', 'hot', 'num_failed_logins', 'logged_in',
    'num_compromised', 'root_shell', 'su_attempted', 'num_root', 'num_file_creations',
    'num_shells', 'num_access_files', 'num_outbound_cmds', 'is_host_login',
    'is_guest_login', 'count', 'srv_count', 'serror_rate', 'srv_serror_rate',
    'rerror_rate', 'srv_rerror_rate', 'same_srv_rate', 'diff_srv_rate',
    'srv_diff_host_rate', 'dst_host_count', 'dst_host_srv_count',
    'dst_host_same_srv_rate', 'dst_host_diff_srv_rate',
    'dst_host_same_src_port_rate', 'dst_host_srv_diff_host_rate',
    'dst_host_serror_rate', 'dst_host_srv_serror_rate', 'dst_host_rerror_rate',
    'dst_host_srv_rerror_rate', 'label', 'difficulty'
]

# Load training data
df_train = pd.read_csv('../data/raw/NSL-KDD/KDDTrain+.txt', names=columns)

# Load test data
df_test = pd.read_csv('../data/raw/NSL-KDD/KDDTest+.txt', names=columns)

print(f"✅ Training set shape: {df_train.shape}")
print(f"✅ Test set shape: {df_test.shape}")
```

```python
# First 5 rows
df_train.head()
```

## 2. Dataset Information

```python
# Basic info
print("=" * 60)
print("DATASET INFO")
print("=" * 60)
print(f"\n📊 Total samples: {len(df_train)}")
print(f"📊 Total features: {len(df_train.columns)}")
print(f"\n📊 Data types:")
print(df_train.dtypes.value_counts())

print(f"\n📊 Missing values: {df_train.isnull().sum().sum()}")
```

```python
# Statistical summary
df_train.describe()
```

## 3. Attack Type Distribution

```python
# Count attack types
attack_counts = df_train['label'].value_counts()

print("=" * 60)
print("ATTACK TYPE DISTRIBUTION")
print("=" * 60)
print(attack_counts)

# Create binary labels (Normal = 0, Attack = 1)
df_train['binary_label'] = df_train['label'].apply(lambda x: 0 if x == 'normal' else 1)

normal_count = (df_train['binary_label'] == 0).sum()
attack_count = (df_train['binary_label'] == 1).sum()

print(f"\n📊 Normal traffic: {normal_count} ({normal_count/len(df_train)*100:.1f}%)")
print(f"📊 Attack traffic: {attack_count} ({attack_count/len(df_train)*100:.1f}%)")
```

```python
# Visualize attack distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Top 15 attack types
top_attacks = attack_counts.head(15)
axes[0].barh(range(len(top_attacks)), top_attacks.values, color='coral')
axes[0].set_yticks(range(len(top_attacks)))
axes[0].set_yticklabels(top_attacks.index)
axes[0].set_xlabel('Count')
axes[0].set_title('Top 15 Attack Types in NSL-KDD', fontsize=12, fontweight='bold')
axes[0].invert_yaxis()

# Plot 2: Normal vs Attack
colors = ['#2ecc71', '#e74c3c']
axes[1].pie([normal_count, attack_count], labels=['Normal', 'Attack'], autopct='%1.1f%%',
            colors=colors, explode=(0.05, 0.05), startangle=90)
axes[1].set_title('Normal vs Attack Traffic Distribution', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.savefig('../docs/eda_attack_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Chart saved: docs/eda_attack_distribution.png")
```

## 4. Categorical Features Analysis

```python
# Analyze categorical features
categorical_cols = ['protocol_type', 'service', 'flag']

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

for i, col in enumerate(categorical_cols):
    value_counts = df_train[col].value_counts()
    axes[i].bar(value_counts.index, value_counts.values, color='steelblue')
    axes[i].set_title(f'{col.upper()} Distribution', fontsize=11, fontweight='bold')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.savefig('../docs/eda_categorical_features.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Chart saved: docs/eda_categorical_features.png")
```

## 5. Feature Correlation Analysis

```python
# Select numerical features for correlation (first 20 numerical columns)
numerical_cols = df_train.select_dtypes(include=[np.number]).columns[:20]
correlation_matrix = df_train[numerical_cols].corr()

# Plot correlation heatmap
plt.figure(figsize=(16, 12))
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
sns.heatmap(correlation_matrix, mask=mask, annot=False, cmap='coolwarm',
            center=0, square=True, linewidths=0.5)
plt.title('Feature Correlation Matrix (Top 20 Features)', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('../docs/eda_correlation_matrix.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Chart saved: docs/eda_correlation_matrix.png")
```

## 6. Class Imbalance Analysis

```python
# Check if class imbalance exists
from collections import Counter

# Count per attack class
class_counts = Counter(df_train['label'])

# Find minority classes
print("=" * 60)
print("CLASS IMBALANCE ANALYSIS")
print("=" * 60)

for attack, count in sorted(class_counts.items(), key=lambda x: x[1]):
    percentage = (count / len(df_train)) * 100
    print(f"{attack:20s}: {count:6d} ({percentage:.2f}%)")

# Identify minority classes (less than 1%)
minority_classes = [(k, v) for k, v in class_counts.items() if (v / len(df_train)) < 0.01]
print(f"\n⚠️ Minority classes (<1%): {len(minority_classes)}")
for cls, cnt in minority_classes:
    print(f"   - {cls}: {cnt} samples")

if len(minority_classes) > 0:
    print("\n✅ Solution: Use SMOTE for class balancing during preprocessing")
```

## 7. Key Insights Summary

```python
print("=" * 60)
print("📋 KEY INSIGHTS SUMMARY")
print("=" * 60)

insights = [
    f"1. Dataset contains {attack_count} attacks ({attack_count/len(df_train)*100:.1f}%) and {normal_count} normal samples",
    f"2. Total attack types: {len(attack_counts)}",
    f"3. Top attack: {attack_counts.index[0]} ({attack_counts.values[0]} samples)",
    f"4. No missing values found - clean dataset",
    f"5. Categorical features: protocol_type ({df_train['protocol_type'].nunique()}), service ({df_train['service'].nunique()}), flag ({df_train['flag'].nunique()})",
    f"6. Class imbalance present - will use SMOTE in preprocessing",
    f"7. Numerical features require normalization (0-1 range)",
    f"8. Real-world attacks include: {', '.join(attack_counts.head(10).index)}"
]

for insight in insights:
    print(insight)
```

## 8. Save Processed Information

```python
# Save insights to a text file
with open('../docs/eda_insights.txt', 'w') as f:
    f.write("CIDRS - NSL-KDD EDA Insights\n")
    f.write("=" * 50 + "\n\n")
    for insight in insights:
        f.write(insight + "\n")
    f.write(f"\n\nAttack type distribution:\n")
    for attack, count in attack_counts.items():
        f.write(f"  {attack}: {count}\n")

print("✅ Insights saved to: docs/eda_insights.txt")
print("\n✅ EDA Complete! Ready for preprocessing (Day 9-10)")
print("📁 Charts saved in docs/ folder:")
print("   - eda_attack_distribution.png")
print("   - eda_categorical_features.png")
print("   - eda_correlation_matrix.png")
```
